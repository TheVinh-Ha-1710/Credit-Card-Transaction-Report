# Datestyle Transaformation

### Why need to format datestyle?
- *PostgreSQL* database does not accept date input with format **"DD-MM-YYYY"**.
- All the columns in our dataset is format as **"DD-MM-YYYY"**, which cannot be ingested.
- Therefore, we need to format our date columns to **"YYYY-MM-DD"**.

```python
import pandas as pd
import numpy as np
```

```python
# Path to our data files
credit_card = "data/credit_card.csv"
cc_add = "data/cc_add.csv"
```

### Datasets overview

```python
df1 = pd.read_csv(credit_card)
df1.head()
```

```python
df2 = pd.read_csv(cc_add)
df2.head()
```

### Data transformation

```python
df1["Week_Start_Date"] = pd.to_datetime(df1["Week_Start_Date"], format="%d-%m-%Y").dt.strftime("%Y-%m-%d")
df1.head()
```

```python
df2["Week_Start_Date"] = pd.to_datetime(df2["Week_Start_Date"], format="%d-%m-%Y").dt.strftime("%Y-%m-%d")
df2.head()
```

### Save the formated datasets 

```python
df1.to_csv(credit_card, index=False)
df2.to_csv(cc_add, index=False)
```
